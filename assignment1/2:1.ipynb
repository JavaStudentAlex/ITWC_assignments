```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 1000)
```

```python
dataset = pd.read_csv("dataset.csv", index_col=0)
```

```python
def plot_order_time_distribution(dataframe): #function to plot dataset
    order_array = dataset.index.values.astype(int)
    time_array = dataset["time_in_seconds"].values.astype("float64")
    plt.plot(order_array, time_array)
    plt.show()
```

```python
plot_order_time_distribution(dataset) #plotting start dataset
```

```python
raws_quantity_to_delete = round(len(dataset)*0.2) #deleting 20% last dataset values
start_delete_index = len(dataset) - 1 - raws_quantity_to_delete
indexes_to_delete = dataset[:-1*raws_quantity_to_delete:-1].index.values
dataset = dataset.drop(indexes_to_delete)
```
